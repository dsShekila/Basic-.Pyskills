### **Merging and Joining Data**

Merging and joining allow you to combine data from multiple sources based on common columns.

For example, if we have two dataframes df1 and df2 cointaining different details of the same patients we can combine them as follows.

```python
df1 = pd.DataFrame({
    'patient_id': [101, 102, 103],
    'name': ['Alice Kariuki', 'Brian Otieno', 'Catherine Njeri'],
    'age': [29, 34, 41],
    'gender': ['F', 'M', 'F']
})

df1
```

```python
# Health record 2: Recent clinical visits
df2 = pd.DataFrame({
    'patient_id': [101, 103, 104],
    'visit_date': ['2025-05-01', '2025-05-03', '2025-05-04'],
    'diagnosis': ['Flu', 'Hypertension', 'Malaria']
})
df2
```

```python
# Merge the two on patient_id
merged_df = pd.merge(df1, df2, on='patient_id', how='outer')
print(merged_df)
```

- pd.merge

    - This functions is used to combine rows from two DataFrames based on a specified condition.
- df1 and df2
    - These represent two separate Pandas DataFrames that you want to merge. Each DataFrame is a tabular data structure with rows and columns containing your data.

- on='common_column'
    - This is the most crucial part. It specifies the key column(s) used to perform the merge.
    - on='common_column' indicates that the merge should be based on a column named 'common_column' that must exist in both DataFrames (df1 and df2).
- How the Merge Works
    - The merge function identifies rows in df1 and df2 that have the same value in the 'common_column'.
    - It then creates a new DataFrame (merged_df) that combines the columns from both df1 and df2 for those matching rows.
    - Rows from df1 and df2 that don't share a common value in the specified column are excluded from the merged DataFrame.

### **Combine**
The combine function combines a dataframe with another dataframe using the  using funtion (either predefined or created functions) to element-wise combine columns. The row and column indexes of the resulting DataFrame will be the union of the two


```python
#Example 1
df1=pd.DataFrame({'A':[1,2,3],'B':[4,5,6]})
df2=pd.DataFrame({'A':[7,8,9],'B':[10,11,12]})
df1.combine(df2,func=lambda x,y:x+y)
```

```python
#Example 2
import numpy as np
df1 = pd.DataFrame({'A': [5, 0], 'B': [2, 4]})
df2 = pd.DataFrame({'A': [1, 1], 'B': [3, 3]})
df1.combine(df2, np.minimum)
```

### **Concat**
The concat method, combines two dataframes. combines  a dataframe based on a certain axis. The dataframe are combined;
1. Vertically (stacking rows on top of each other)
2. Horizontally (joining columns side by side)




```python
# example
df1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df2 = pd.DataFrame({'A': [5, 6], 'B': [7, 8]})

vertical_concat = pd.concat([df1, df2], axis=0)
horizontal_concat = pd.concat([df1, df2], axis=1)
print("Vertical:")
display(vertical_concat)
print("Horizontal:")
display(horizontal_concat)
```

### Join

Joins the dataframes based on the index or a particular column

```python
## Example

df = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3', 'K4', 'K5'],
                   'A': ['A0', 'A1', 'A2', 'A3', 'A4', 'A5']})
other = pd.DataFrame({'key': ['K0', 'K1', 'K2'],
                      'B': ['B0', 'B1', 'B2']})

df.join(other, lsuffix='_caller', rsuffix='_other')
```

If we want to join using the key columns, we need to set key to be the index in both df and other. The joined DataFrame will have key as its index.

```python
df.set_index('key').join(other.set_index('key'))
```
